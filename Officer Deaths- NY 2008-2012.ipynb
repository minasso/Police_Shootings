```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
from pylab import rcParams
rcParams['figure.figsize'] = 14,6
```

```python
from functions import pulldata
df = pulldata('select person, canine, cause_short as cause, year, dept_name as dept, state from officers where year between 2008 and 2012 and state= "NY"')
```

```python
ax = sns.countplot(x="cause", data=df,order = df['cause'].value_counts().index)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
```

Focus attention on those causes in which the officer was intentionally killed

```python
killed = ['Gunfire','Vehicular assault','Vehicle pursuit','Assault','Stabbed']
```

```python
dk = df[df['cause'].isin(killed)].copy()
dk.sort_values('cause',inplace=True)
dk.reset_index(drop=True,inplace=True)
sns.countplot(dk.cause,order=dk['cause'].value_counts().index)
```
