```python
def map_plotter(df):
    data = dict(
               type = 'scattergeo',
               locationmode = 'USA-states',
               lon = df.longitude,
               lat = df.latitude,
               text =df.city,
               mode = 'markers',
               name='2015-2016',
               hoverinfo = 'text+name',
               marker = dict(
                   size = 5,
                   opacity = 0.95,
                   color = 'rgb(240, 140, 45)')
               )

    layout = dict(
             title = 'Fatal Police Shootings: US',
             geo = dict(
                 scope = 'usa',
                 projection = dict(type = 'albers usa'),
                 showland = True,
                 landcolor = 'rgb(250, 250, 250)',
                 subunitwidth = 1,
                 subunitcolor = 'rgb(215, 215, 215)',
                 countrywidth = 1,
                 countrycolor = 'rgb(215, 215, 215)',
                 showlakes = True,
                 lakecolor = 'rgb(255, 255, 255)')
             )
    figure = go.Figure(data = [data], layout = layout)
    iplot(figure)
map_plotter(df)
```
